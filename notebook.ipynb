You have been asked to support an investment firm by analyzing trends in high-growth companies. They are interested in understanding which industries are producing the highest valuations and the rate at which new high-value companies are emerging. Providing them with this information gives them a competitive insight as to industry trends and how they should structure their portfolio looking forward.

The task for this unguided SQL project was to identify the trends in the three best-performing industries based on the number of new unicorns created over the last three years (2019, 2020, and 2021) combined.

The following SQL query was implemented and the top performing industries were:
1. E-commerce & direct-to-consumer
2. Fintech
3. Internet software and services

```python
-- Creating a CTE called TOP3 to identiify the top 3 industries for the number of unicorns created from 2019-2021 combined
WITH TOP3 AS (SELECT industry,
       COUNT(*) AS num_unicorns
FROM industries
INNER JOIN dates
ON industries.company_id = dates.company_id
WHERE EXTRACT(year FROM date_joined) BETWEEN 2019 AND 2021
GROUP BY industry
ORDER BY num_unicorns DESC
LIMIT 3)

-- Querying the tables for the top 3 industries identified in the TOP3 CTE
SELECT industry,
EXTRACT(year FROM date_joined) AS year,
COUNT(*) AS num_unicorns,
ROUND(AVG(valuation / 1000000000), 2) AS average_valuation_billions  --Calculating average valuation in billions
FROM industries
INNER JOIN dates
ON industries.company_id = dates.company_id
INNER JOIN funding
ON dates.company_id = funding.company_id
WHERE industry in (SELECT industry FROM TOP3)  -- Selecting only the top 3 industries identified in the CTE
AND EXTRACT(year FROM date_joined) BETWEEN 2019 AND 2021
GROUP BY industry, year
ORDER BY industry, year DESC; -- Displaying the trends for the top 3 industries
```
